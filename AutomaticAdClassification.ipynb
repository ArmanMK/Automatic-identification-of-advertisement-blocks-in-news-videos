```python
import time
start = time.time()

#################################################### Data Acquisition
import sklearn.datasets
from sklearn.datasets import load_svmlight_file
import numpy as np


TrainData=sklearn.datasets.load_svmlight_file('train.txt',multilabel=True)
XTrain=TrainData[0]
XTrain=XTrain.todense()

YTrain=TrainData[1]
YTrain= np.asarray(YTrain)


TestData=sklearn.datasets.load_svmlight_file('test.txt',multilabel=True)
XTest=TestData[0]
XTest=XTest.todense()

YTest=TestData[1]
YTest=np.asarray(YTest)


KaggleData=sklearn.datasets.load_svmlight_file('kaggle.txt',multilabel=True)
XKaggle=KaggleData[0]
XKaggle=XKaggle.todense()


DataAqTime= time.time()-start 
######################################################## KNN Classifier
start = time.time()
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection  import cross_val_score
from sklearn.metrics import accuracy_score

#### Creating the KNN model using the training data
knn=KNeighborsClassifier()
knn.fit(XTrain, YTrain.ravel())

KNNFitTime=time.time()- start 
print(f'It took {KNNFitTime} seconds for the KNN Classifier to fit')

start=time.time() 

#### predicting classes of test data and finding the accuracy score of KNN
KNNPrediction=knn.predict(XTest)
KNNAccuracy = accuracy_score(YTest, KNNPrediction)
print(f'Mean accuracy score for KNN: {KNNAccuracy:.3}')

KNNTime=time.time()- start

print(f'It took {KNNTime} seconds for the KNN Classifier to predict.')
######################################################## Random Forest Classifier
start = time.time()
from sklearn.ensemble import RandomForestClassifier

#### Creating the Random Forest model using the training data
rf = RandomForestClassifier()
rf.fit(XTrain, YTrain.ravel())

RandomForestFitTime=time.time()-start
print(f'It took {RandomForestFitTime} seconds for the Random Forest Classifier to fit')
start = time.time()

#### predicting classes of test data and finding the accuracy score of Random Forest

RandomForestPrediction = rf.predict(XTest)
RFAccuracy = accuracy_score(YTest, RandomForestPrediction)
print(f'Mean accuracy score for Random Forest: {RFAccuracy:.3}')


RandomForestTime=time.time()-start

print(f'It took {RandomForestTime} seconds for the Random Forest Classifier to predict')
####################################################### SVM Classifier
########## F0r SVM classifier to perform the best, the data needs to be scaled into a range
###### After rescaling the data, my accuracy score increased dramatically
from sklearn import preprocessing
XXTrain = preprocessing.scale(XTrain)
XXTest = preprocessing.scale(XTest)
XXKaggle = preprocessing.scale(XKaggle)



start = time.time()
from sklearn import svm
#### Creating the SVM model using the training data

SupportVectorMachine = svm.SVC()
SupportVectorMachine.fit(XXTrain,YTrain.ravel())

SVMFitTime=time.time()-start
print(f'It took {SVMFitTime} seconds for the SVM Classifier to fit')
start = time.time()

#### predicting classes of test data and finding the accuracy score of SVM

SVMPrediction = SupportVectorMachine.predict(XXTest)
SVMAccuracy = accuracy_score(YTest, SVMPrediction)
print(f'Mean accuracy score for SVM: {SVMAccuracy:.3}')



SVMTime=time.time()-start
print(f'It took {SVMTime} seconds for the SVM Classifier to predict')
###################################################### Neural Networks Classifier
start = time.time()
### also, for multi-layer perceptron, the preprocessing helps a lot in model fitting time
from sklearn import preprocessing
XXTrain = preprocessing.scale(XTrain)
XXTest = preprocessing.scale(XTest)
XXKaggle = preprocessing.scale(XKaggle)
from sklearn.neural_network import MLPClassifier

#### Creating the MLP model using the training data

mlp = MLPClassifier(solver='sgd',hidden_layer_sizes=(100,20,10),activation='relu',learning_rate='adaptive')
mlp.fit(XXTrain, YTrain.ravel())             

NNFitTime=time.time()-start
print(f'It took {NNFitTime} seconds for the Neural Netwroks Classifier to fit')
start = time.time()
           
#### predicting classes of test data and finding the accuracy score of Multi-Layer Perceptron

MLPPrediction = mlp.predict(XXTest)
MLPAccuracy = accuracy_score(YTest, MLPPrediction)
print(f'Mean accuracy score for NN: {MLPAccuracy:.3}')

NNTime=time.time()-start
print(f'It took {NNTime} seconds for the Neural Netwroks Classifier to predict')

################################################################




```
